# Exercise 4

Design a stack that supports push, pop, and a get_min() operation that returns the minimum
element in the stack — all in O(1) time.


The challenge: when you pop an element, the minimum might change. Think about how to keep
track of the current minimum at every level of the stack.
Constraints


All get_min(), push(), and pop() must run in O(1) time.

```python
class Min_Stack():
    '''Class stack LIFO big O of 1'''

    def __init__(self, arr = None):
        '''Constructor'''

        self.arr = [] if arr is None else arr
        self.min = []

    def push(self, value: int):
        '''Push value to the top of the array'''
        self.arr.append(value)
        if self.min == [] or value < self.min[-1]: 
            self.min.append(value)
        
    def peek(self):
        '''Peek top of the stack'''
        return self.arr[-1]
    
    def pop(self):
        '''returns and give top of stack'''
        if self.arr[-1] == self.min[-1]: self.min.pop()

        return self.arr.pop()
    
    
    def get_min(self):
        '''return min of stack'''
        return self.min[-1]
```

```python
#unit testing
ms = Min_Stack()
ms.push(5)
ms.push(3)
ms.push(7)


print(ms.get_min())
ms.pop()
ms.pop()
print(ms.get_min())
```
