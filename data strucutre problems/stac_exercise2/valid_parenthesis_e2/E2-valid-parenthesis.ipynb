# Exercises
now that we can access to our stack class, lets start with the exercises 

-Given a string containing only '(', ')', '{', '}', '[', and ']', determine if the input string is
valid.

-A string is valid if every opening bracket has a corresponding closing bracket of the same type, and
brackets are closed in the correct order (most recently opened closes first).

```python
def valid_parentheses(arr: list) -> bool:
    '''Returns true or false if valid parentheses sequence'''

    stack = Stack() #Create stack
    correct = [] #Create list to append results
    opening = ['(', "[", "{", ] #list of openings
    for element in arr: #loop through element of array
        if element in opening: #check if the opening is corrrect if so add to stack
            stack.push(element)
            continue
        
        value = stack.pop() #get las value and pop it
        #Check if valid closure
        if value == '(' and element == ')': correct.append(True)
        elif value == '[' and element == ']': correct.append(True)
        elif value == '{' and element == '}': correct.append(True)
        else: correct.append(False) #not correct parenthesis
    
    return all(element is True for element in correct)
```

```python
#Unit testing
print(valid_parentheses('()'))
```
