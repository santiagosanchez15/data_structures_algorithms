# Exercise 7
Given an array temperatures, return an array answer where answer[i] is the number of days
to wait after day i to get a warmer temperature. 

If no warmer day exists, set answer[i] = 0.
This problem introduces the monotonic stack pattern — a stack kept in sorted order to efficiently
find the next greater element.

A monotonic stack is a specialized data structure, maintainin elements in increasing and decresing order, 
solvin the next greater and smaller element

```python
def daily_temperatures(arr: list) -> list:
    '''Returns a lis of next warmer days from current day'''

    stack = [] #initialize the stack
    answer = [0] * len(arr) #create a fixed size  list with value of 0

    for  i in range(len(arr)): #loop through the indexes
        
        while stack and arr[i] > arr[stack[-1]]: #if the there's a value in the stack and the element at index i is greater than the value in my stack, then start

                answer[stack[-1]] = i - stack[-1] #append the value at given index, get the difference 
                stack.pop() #take off the value from the stack
            
        stack.append(i) #append the index to the stack so it can later be checked
        
    return answer
```

```python
#Unit testing
temps = [55,38,53,81,61,93,97,32,43,78]
print(daily_temperatures(temps))
```
