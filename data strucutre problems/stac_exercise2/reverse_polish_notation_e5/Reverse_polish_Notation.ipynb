# Exercise 5 polish notation
In RPN, operands come before the operator: 3 4 + means 3 + 4 = 7. 

When you encounter an operator, apply it to the two most recently seen numbers. Division truncates toward zero.


Constraints
Valid operators are +, -, *, /. The expression is always valid.

```python
def polish_notation(arr: list) -> int:
    '''Resolves equation through polish notation'''

    stack = []
    for element in arr:

        if element == '+': 
            val1 = int(stack.pop())
            val2 = int(stack.pop())
            stack.append(val1 + val2)
        elif element == "-":
            val1 = int(stack.pop())
            val2 = int(stack.pop())
            stack.append(val2 - val1)
        elif element == "*":
            val1 = int(stack.pop())
            val2 = int(stack.pop())
            stack.append(val1 * val2)
        elif element == "/":
            val1 = int(stack.pop())
            val2 = int(stack.pop())
            stack.append(val2 // val1)
        else:
            stack.append(int(element))
        
    return stack[0]


        
```

```python
#unit testing
tokens = [['4','13','5','/','+'], ['2','1','+','3','*']]
for token in tokens:
    print(polish_notation(token))
```
